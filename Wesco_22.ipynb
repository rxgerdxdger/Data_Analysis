```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from matplotlib.gridspec import GridSpec
from IPython.display import display
from IPython.display import HTML
from ipywidgets import Textarea
```

```python
wesco = pd.read_csv("2022 Weslaco Gin study data with stats.csv")

```

```python
grouped_Entry = wesco.groupby("ENTRY")
seedcotton_weight = grouped_Entry["Seedcotton Weight (g)"]
mean_seedcotton_weight = seedcotton_weight.mean()
scw_std = seedcotton_weight.std()
```

```python
fiber_weight = grouped_Entry["Fiber Wt (g)"]
mean_fiber_weight = fiber_weight.mean()

fib_std = fiber_weight.std()
```

```python
mean_seedcotton_weight.sort_values(ascending = False).rank(method = 'dense', ascending = False)



```

```python
gs = GridSpec(2,2)
fig = plt.figure()

ax1 = plt.subplot(gs[0, 0])


ax1 = plt.subplot(gs[0, 0])
ax2 = plt.subplot(gs[0, 1])
ax3 = plt.subplot(gs[1, 0])
ax4 = plt.subplot(gs[1, 1])

plt.subplots_adjust(left =0.1, bottom = 0.9, right = 4.5, top =3.0, wspace=0.1, hspace=0.1)
mean_seedcotton_weight.plot(kind='pie', x='x',y='y', ax = ax4)
ax2.bar(grouped_Entry.groups.keys(), mean_seedcotton_weight)
plt.title('Average & Standard Deviation of SeedCotton Weight')
plt.xlabel("")
plt.ylabel(")")

# Create a second bar plot of the standard deviation of the seedcotton weight
ax2.bar(grouped_Entry.groups.keys(), scw_std, bottom=mean_seedcotton_weight)
mean_seedcotton_weight_rank = mean_seedcotton_weight.sort_values(ascending = False).rank(method = 'dense', ascending = True)

ax1.bar(mean_seedcotton_weight_rank.index, mean_seedcotton_weight_rank)
ax1.set_title('Mean Seedcotton Weight Rank')
ax1.set_xlabel('Entry')
ax1.set_ylabel('Rank')

# Add a legend
ax2.legend(["Mean", "Standard Deviation"], loc="lower left")

# Show the plot
ax3.scatter(grouped_Entry.groups.keys(), mean_seedcotton_weight, s=scw_std*20)
ax3.set_xlabel("Rep")
ax3.set_ylabel("Mean Seedcotton Weight (g)")
ax3.set_title("Average & Standard Deviation of SeedCotton Weight")
# Add a legend
ax3.legend(["Standard Deviation(Size Variation attributed to deviation)"], loc = 'lower left',bbox_to_anchor=(0.00, 0.00))
plt.xticks(rotation=90)
# Show the plot
plt.show()
# 
plt.show()
```
